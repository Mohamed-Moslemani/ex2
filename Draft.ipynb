```python
import pandas as pd
import numpy as np 
from scipy.stats import kurtosis, skew
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv("data/Retail_Dataset.csv")
```

```python
df.head()
```

```python
def summarize_independent_vars(df, target_col=None):
    X=df.copy()
    for col in ['week', target_col]:
        if col and col in X.columns:
            X.drop(columns=col, inplace=True)
    X= X.select_dtypes(include='number')
    summary = X.describe(percentiles=[.25, .5, .75]).T   
    summary['skew']= X.apply(skew,     nan_policy='omit')
    summary['kurtosis']= X.apply(kurtosis, nan_policy='omit')
    return summary
```

```python
stats_df=summarize_independent_vars(df, target_col='weekly_revenue_log')
(stats_df.head())
```
